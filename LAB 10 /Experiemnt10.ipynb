```python
import pandas as pd
from itertools import combinations

# Load the dataset
data = pd.read_csv('/content/Market_Basket_Optimisation.csv', header=None)

# Convert the data to a list of transactions
transactions = data.apply(lambda row: row.dropna().tolist(), axis=1).tolist()

# Step 1: Generate frequent itemsets (Custom Apriori implementation)
def create_candidates(itemsets, length):
    candidates = set()
    for itemset in itemsets:
        for item in items:
            candidate = itemset.union([item])
            if len(candidate) == length:
                candidates.add(candidate)
    return candidates

def calculate_support(transactions, itemsets):
    itemset_counts = {itemset: 0 for itemset in itemsets}
    for transaction in transactions:
        transaction = set(transaction)
        for itemset in itemsets:
            if itemset.issubset(transaction):
                itemset_counts[itemset] += 1
    return {itemset: count / len(transactions) for itemset, count in itemset_counts.items()}

def apriori(transactions, min_support=0.01):
    items = {item for transaction in transactions for item in transaction}
    frequent_itemsets = []
    current_itemsets = [{item} for item in items]

    # Iteratively generate itemsets of increasing length
    length = 1
    while current_itemsets:
        itemset_support = calculate_support(transactions, current_itemsets)
        current_itemsets = [itemset for itemset, support in itemset_support.items() if support >= min_support]
        frequent_itemsets.extend([(itemset, itemset_support[itemset]) for itemset in current_itemsets])
        length += 1
        current_itemsets = create_candidates(current_itemsets, length)

    return frequent_itemsets

# Run Apriori algorithm with minimum support of 0.01
frequent_itemsets = apriori(transactions, min_support=0.01)

# Step 2: Extract association rules
def generate_rules(frequent_itemsets, min_confidence=0.3):
    rules = []
    for itemset, support in frequent_itemsets:
        if len(itemset) > 1:
            for consequent in itemset:
                antecedent = itemset - {consequent}
                antecedent_support = next(s for i, s in frequent_itemsets if i == antecedent)
                confidence = support / antecedent_support
                if confidence >= min_confidence:
                    lift = confidence / (next(s for i, s in frequent_itemsets if i == {consequent}))
                    rules.append((antecedent, {consequent}, support, confidence, lift))
    return rules

# Generate rules with a minimum confidence of 0.3
association_rules = generate_rules(frequent_itemsets, min_confidence=0.3)

# Step 3: Analyze the results
# Sort rules by lift and display top 10
association_rules = sorted(association_rules, key=lambda x: x[4], reverse=True)
print("Top 10 Association Rules:")
for antecedent, consequent, support, confidence, lift in association_rules[:10]:
    print(f"Rule: {antecedent} -> {consequent}")
    print(f"Support: {support:.2f}, Confidence: {confidence:.2f}, Lift: {lift:.2f}\n")
```
