```python
#performing the normilization at the age
mean = data["Age"].mean()
max_value = data["Age"].max()
min_value = data["Age"].min()

print("mean : " , mean)
print("max values " , max_value)
print("min values" , min_value)
```

```python
#find the min-max algorithum
#find the differnce
diff = max_value - min_value
minmax = (data['Age']-min_value ) / diff
print(minmax)
```

```python
#now find the z score normlization
#find the standerd deviation
standerd = data['Age'].std()
print(standerd)
```

```python
#find the z score scalling
z_score = (data['Age']-mean)/standerd
print(z_score)
```

```python
import seaborn as sc
sc.distplot(z_score)
```
