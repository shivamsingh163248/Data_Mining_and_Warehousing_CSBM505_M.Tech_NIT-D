```python
#importing the librerys
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#now reading the file using the pandas
dataset = pd.read_csv("Loan dataset train.csv")
```

```python
#print the dataset head 20head info and describe
print(dataset)
```

```python
dataset.mean(numeric_only=True)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
#printing the historgram using matplotlib
plt.hist(dataset["ApplicantIncome"] , color='red' )
#writing the code for the y and x axis
plt.ylabel('ApplicantIncome')
plt.xlabel('friquency')
#also the set the title of the plot
plt.title("income historgram")
plt.show()
```

```python
#using the seabor we are the doing the count plot
sns.boxplot(dataset['ApplicantIncome'])
plt.show()
```

```python
#creating the count plot using the secborn
sns.countplot( data=dataset , x='Property_Area' ) #also select the attribute of the aresas
#set title of the count plot
plt.title("count plot ")
plt.show()
```

```python
#now learning the bibrent analysis
#learning the scatter plot
sns.scatterplot(data= dataset )
plt.title("scatter plot ")
plt.show()
```

```python
#now learning the bibrent analysis
#learning the scatter plot
sns.scatterplot(data= dataset , x='Property_Area'  , y = 'ApplicantIncome')

plt.title("scatter plot ")
plt.show()
```

```python
#creating the violen plot
sns.violinplot(data = dataset , x='Property_Area'  , y = 'ApplicantIncome')
plt.title("violinpolt")
plt.show()     #creating the x lavel and y lavel
```
